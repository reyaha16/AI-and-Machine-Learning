```python
def unit_converter(type, value, option):
  """
    Converts units between length, weight, and volume.

    Parameters:
    conversion_type (str): Type of conversion ('length', 'weight', 'volume')
    value (float): Value to convert
    option (int): Conversion option number

    Returns:
    float: Converted value
    """

  if  type == 'length':
      if option == 1:
        return value * 3.28084   # meters to feet
      elif option == 2:
        return value / 3.28084   # feet to meters


  elif type == "weight":
      if option == 1:
        return value * 2.20462   # kg to lbs
      elif option == 2:
        return value / 2.20462   # lbs to kg

  elif type == "volume":
      if option == 1:
          return value * 0.264172  # liters to gallons
      elif option == 2:
          return value / 0.264172  # gallons to liters

  else:
        print("Unsupported conversion type.")
try:
    print("Choose conversion type: length / weight / volume")
    type = input("Enter type: ").lower()

    if type =='length':
        print("Choose option:")
        print("1. Meters to Feet")
        print("2. Feet to Meters")
        option = int(input("Enter option (1 or 2): "))

    elif type == 'weight':
        print("Choose option:")
        print("1. Kilograms to Pounds")
        print("2. Pounds to Kilograms")
        option = int(input("Enter option (1 or 2): "))

    elif type == 'volume':
        print("Choose option:")
        print("1. Liters to Gallons")
        print("2. Gallons to Liters")
        option = int(input("Enter option (1 or 2): "))

    else:
        print("Unsupported conversion type.")
        exit()


    value = float(input("Enter value to convert: "))

    result = unit_converter(type, value, option)
    print("Converted value:", result)

except ValueError as e:
    print("Error:", e)
except Exception:
    print("Invalid input. Please try again.")


```

### Create a Python program that performs various mathematical operations on a list of numbers.

```python
def calculate_sum(numbers):
    """
    Returns the sum of a list of numbers.
    """
    return sum(numbers)

def calculate_average(numbers):
    """
    Returns the average of a list of numbers.
    """
    return sum(numbers) / len(numbers)

def find_max(numbers):
    """
    Returns the maximum value in a list.
    """
    return max(numbers)

def find_min(numbers):
    """
    Returns the minimum value in a list.
    """
    return min(numbers)
try:
    print("Choose operation: sum / average / max / min")
    operation = input("Enter operation: ").lower()

    numbers = list(map(float, input("Enter numbers separated by space: ").split()))

    if len(numbers) == 0:
        raise ValueError("List cannot be empty.")

    if operation == "sum":
        print("Result:", calculate_sum(numbers))
    elif operation == "average":
        print("Result:", calculate_average(numbers))
    elif operation == "max":
        print("Result:", find_max(numbers))
    elif operation == "min":
        print("Result:", find_min(numbers))
    else:
        print("Invalid operation.")

except ValueError:
    print("Invalid input. Please enter valid numbers.")
```
